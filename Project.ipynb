```python
import cloudscraper

scraper = cloudscraper.create_scraper(
    browser={
        "browser": "chrome",
        "platform": "windows",
        "mobile": False})
url = "https://sofifa.com/players"
response = scraper.get(url)
print(response.status_code)
print(response.text[:500])
```

```python
from bs4 import BeautifulSoup
```

```python
soup=BeautifulSoup(response.text, 'html')
```

```python
table=soup.find_all('table')[0]
```

```python
header=table.find_all('th')
```

```python
title=[head.text.strip() for head in header]
```

```python
import pandas as pd
df=pd.DataFrame(columns=title)
df
```

```python
column= table.find_all('tr')
for co in column[1:]:
    data=co.find_all('td')
    dat=[d.text.strip() for d in data]
    length=len(df)
    df.loc[length]=dat
```

```python
df
```

```python
df.to_csv("sofifa_players.csv", index=False)
df_csv = pd.read_csv("sofifa_players.csv")
print(df_csv.head())
```
